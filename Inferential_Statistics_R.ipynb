ONE SAMPLE T TEST

We are checking wether the mean of the variable (fixed acidity) in quaestion is equal to the specified value

Null hypothesis : The fixed mean of acidity = 6.10

Alternative Hypothesis: The mean is not = 6.10

```r
t.test(df$fixed.acidity, mu=6.10)
```

p value is very small meaning we reject Null hypothesis and conclude that The mean is not = 6.10

Confidence Interval

```r
t.test(df$fixed.acidity, mu=6.10, conf.level = 0.01) #99% CONFIDENCE INTERVAL
```

```r
t.test(df$fixed.acidity, mu=6.10, conf.level = 0.25) # 25 % CONFIDENCE INTERVAL
```

```r
t.test(df$fixed.acidity, mu=6.10, alternative='less') #99% CONFIDENCE INTERVAL
```

```r
t.test(df$fixed.acidity, mu=6.10, conf.level = 0.01) #99% CONFIDENCE INTERVAL
t.test(df$fixed.acidity, mu=6.10, alternative = 'less') #99% CONFIDENCE INTERVAL
t.test(df$fixed.acidity, mu=6.10, alternative = 'greater') #99% CONFIDENCE INTERVAL
```

PAIRED T TEST

```r
summary(df)
```

```r
head(df,3)
```

Null hypothesis = true difference in means is = 0 # There is no difference


Anternative hypothesis = true difference in means is not equal to 0 # There is significant difference

```r
t.test(df$fixed.acidity,df$residual.sugar, paired=TRUE)
```

Conclusion p value is very small reject Null hypothesis and conclude that true difference in means is not equal to 0

```r
t.test(df$fixed.acidity,df$residual.sugar, paired=TRUE, conf.level=0.1)
t.test(df$fixed.acidity,df$residual.sugar, paired=TRUE, conf.level= 0.05)
t.test(df$fixed.acidity,df$residual.sugar, paired=TRUE, conf.level= 0.01)
```

ANOVA ANALYSIS ANALYSING DOFFERENCE IN DIFFERENT ALCOHOL QUALITIES 

```r
anova_model= aov(alcohol~quality,data=df)
```

```r
summary(anova_model)
```

p value very small we reject null hypothesis and conclude that the qualities of different alcohole are different

NOW WE WANT TO TEST AND SEE WHICH EXACTLY OF THE QUALITY LEVELS OF WINE ARE DIFFERENT

WE USE THE PAIRWISE TEST

```r
pairwise.t.test(df$alcohol,df$quality,adjust.method='bonf')
```

```r
# Ensure quality is a factor
df$quality <- as.factor(df$quality)
```

```r
# Perform Tukey's HSD test
tukey_result <- TukeyHSD(anova_model)
```

```r
# Print the Tukey results
print(tukey_result)
```

```r
# Visualizing the Tukey HSD results
plot(tukey_result, las = 2)
```

```r
# Boxplot to visualize differences
library(ggplot2)
ggplot(df, aes(x = quality, y = alcohol)) +
  geom_boxplot(fill = "lightblue", color = "black") +
  geom_jitter(width = 0.2, alpha = 0.5, color = "blue") +
  theme_minimal() +
  labs(title = "Alcohol Content by Quality Level", x = "Quality", y = "Alcohol Content")
```
